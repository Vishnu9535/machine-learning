```python
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
from scipy.stats import norm
```

```python
data = np.arange(1,10)
pdf = norm.pdf(data, 10 ,0.1)
sb.set_style("whitegrid")
sb.lineplot(x=data, y=pdf  ,color = 'black')
plt.xlabel("height")
plt.ylabel("probability density")
```

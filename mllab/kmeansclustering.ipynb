```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans  
df = pd.read_csv('wine-clustering.csv')
df.describe
```

```python
print(df.columns)
selected_features = df[['OD280','Alcohol']]
KMeans_obj = KMeans(n_clusters=3,random_state=42)
KMeans_obj.fit(selected_features)
KMeans_obj.predict(selected_features)
y_kmeans = KMeans_obj.predict(selected_features)
print(y_kmeans)
df.describe
```

```python
centers=KMeans_obj.cluster_centers_
print(centers)
```

```python
import matplotlib.pyplot as plt
plt.scatter(selected_features['OD280'],selected_features['Alcohol'],c=y_kmeans,s=50,cmap='viridis')
plt.scatter(KMeans_obj.cluster_centers_[:,0],KMeans_obj.cluster_centers_[:,1],c='black',s=200,alpha=0.5)
plt.xlabel('OD280')
plt.ylabel('Alcohol')
```

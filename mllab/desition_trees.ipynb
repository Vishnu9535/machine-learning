```python
df['workclass'] =  df['workclass'].replace('?',np.nan)
df['occupation'] =  df['occupation'].replace('?',np.nan)
df['native-country'] =  df['native-country'].replace('?',np.nan)
df.isin(['?']).sum()

```

```python
df.dropna(how = 'any',inplace=True)
df.drop_duplicates()
```

```python
df.shape
df.drop(['fnlwgt','educational-num','marital-status','relationship','race'],axis = 1,inplace =True)
```

```python
df.columns
```

```python
x= df.loc[:,['age','workclass', 'education', 'occupation', 'gender', 'capital-gain',
       'capital-loss', 'hours-per-week', 'native-country']]
y = df.loc[:,['income']]
```

```python
y.head
```

```python
from sklearn.preprocessing import LabelEncoder
y=LabelEncoder().fit_transform(y)
y = pd.DataFrame(y)
```

```python
y.head()
```

```python
numerical_features = x.select_dtypes('number')
categorical_features = x.select_dtypes('object')
print(numerical_features)
```

```python
print(categorical_features)
```

```python
convert_catogorical_features = pd.get_dummies(categorical_features)
convert_catogorical_features.head()
```

```python
all_features = [convert_catogorical_features,numerical_features]
newX = pd.concat(all_features,axis = 1, join = 'inner')
newX.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(newX,y,test_size = 0.33,random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth = 5)
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
print(y_pred)
print(y_pred.shape)
print(y_test.shape)
```

```python
prediction_df = pd.DataFrame()
prediction_df['predicted_salary_class'] = y_pred
prediction_df['actual_salary_class'] = y_test[0].values
print(prediction_df)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plt.figure(figsize = (20,20))
plot_tree(clf,filled = True,fontsize=10)
plt.title("desicision tree trained")
plt.show()
```
